```python
cities = geopandas.read_file("https://github.com/jorisvandenbossche/geopandas-tutorial/blob/main/data/ne_110m_populated_places.zip?raw=true") #carico le città 
cities
```

```python
#come facciamo a capire lo stato di appartenenza di una città
```

```python
#edgecolor: colore bordo; facecolor: colore dello sfondo; figsize: dimensione della figura
ax = countries.plot(edgecolor='k', facecolor='none', figsize=(15, 10))
cities.plot(ax=ax, color='red') #stampa le città sullo stesso grafico creato per i countries
ax.set(xlim=(-20, 60), ylim=(-40, 40)) #settiamo delle limitazioni sugli assi delle X(longitudine) e Y(latitudine)
```

```python
#vedere solo l'italia
ax = countries.plot(edgecolor='k', facecolor='none', figsize=(15, 10))
cities.plot(ax=ax, color='red')
ax.set(xlim=(5, 20), ylim=(36, 48))
italia = countries[countries["name"] == "Italy"].plot()
italia
```

```python
stations = geopandas.read_file("https://github.com/jorisvandenbossche/geopandas-tutorial/blob/main/data/paris_bike_stations_mercator.gpkg?raw=true")
stations
```

```python
stations.plot()
#1e6 è la scala: non sono più gradi ma mentri (in questo caso milioni di metri)
```

```python
!pip install contextily
```

```python
import contextily #serve per importare le mappe su cui visualizzare i dati
```

```python
#visualizzare la mappa con sopra i dati 
ax = stations.plot(figsize=(12,6), markersize=5) #markersize è la grandezza dei pallini
contextily.add_basemap(ax) #aggiungo i pallini alla mappa di base
```

```python
stations['bike_stands'].hist()
#asse delle ascisse: il numero di posti per le bici
#asse delle ordinate: il numero di stazioni con quel numero di posti per le bici
```

```python
ax = stations.plot(figsize=(12, 6), column='available_bikes', legend=True) #mostro i pallini delle bici disponibili con i colori che si differenziano in base al numero 
```

```python
#vedere i pallini colorati sulla mappa
ax = stations.plot(figsize=(24, 12), column='available_bikes', legend=True, markersize=9)
contextily.add_basemap(ax)
```

```python
districts = geopandas.read_file("https://raw.githubusercontent.com/jorisvandenbossche/geopandas-tutorial/main/data/paris_districts_utm.geojson")
districts
```

```python
districts.plot()
```

```python
ax = districts.plot(figsize=(12,6), edgecolor= "k", facecolor = "none") #mostro solo i contorni dei quartieri
contextily.add_basemap(ax)
```

```python
districts.geometry.area #calcolo l'area dei quartieri
```

```python
districts["area"] = districts.geometry.area / 1e6  #aggiungo al geoDataFrame la colonna area (la mostro in kilometri quadrati)
districts
```

```python
districts["population_density"] = districts["population"] / districts.geometry.area * 10**6
districts.plot(column = "population_density", figsize = (12, 6), legend = True)
```
